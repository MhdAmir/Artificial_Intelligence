## Class GRE

```python
class GRE:

  titik1=0
  titik2=800
  titik3=1200
  titik4=1800

  def __init__(self, gre):
    self.gre = gre
    print("self gre =", self.gre)


  def low(self):
    if(self.gre >= self.titik1 and self.gre <= self.titik2):
      return float(1.0)
    elif (self.gre > self.titik2 and self.gre < self.titik3):
      return round(float(self.titik3 - self.gre)/(self.titik3 - self.titik2), 2)
    else:
      return float(0.0)

  def medium(self):
    if (self.gre > self.titik2 and self.gre < self.titik3):
      return round(float(self.gre - self.titik2)/(self.titik3 - self.titik2), 2)
    elif (self.gre > self.titik3 and self.gre < self.titik4):
      return round(float(self.titik4 - self.gre)/(self.titik4 - self.titik3), 2)
    else:
      return float(0.0)

  def high(self):
    if (self.gre > self.titik3 and self.gre < self.titik4):
      return round(float(self.gre - self.titik3)/(self.titik4 - self.titik3), 2)
    elif (self.gre >= self.titik4):
      return float(1.0)
    else:
      return float(0.0)
```

```python
class Centroid_Method:
  titik1=0
  titik2=60
  titik3=70
  titik4=80
  titik5=90
  titik6=100

  def __init__(self, mid_score):
    self.mid_score = mid_score
    print("self mid_score =", self.mid_score)

  def poor(self):
    if(self.mid_score >= self.titik1 and self.mid_score <= self.titik2):
      return float(1.0)
    elif (self.mid_score > self.titik2 and self.mid_score < self.titik3):
      return round(float(self.titik3 - self.mid_score)/(self.titik3 - self.titik2), 2)
    else:
      return float(0.0)

  def fair(self):
    if (self.mid_score > self.titik2 and self.mid_score < self.titik3):
      return round(float(self.mid_score - self.titik2)/(self.titik3 - self.titik2), 2)
    elif (self.mid_score > self.titik3 and self.mid_score < self.titik4):
      return round(float(self.titik4 - self.mid_score)/(self.titik4 - self.titik3), 2)
    else:
      return float(0.0)

  def good(self):
    if (self.mid_score > self.titik3 and self.mid_score < self.titik4):
      return round(float(self.mid_score - self.titik3)/(self.titik4 - self.titik3), 2)
    elif (self.mid_score > self.titik4 and self.mid_score < self.titik5):
      return round(float(self.titik5 - self.mid_score)/(self.titik5 - self.titik4), 2)
    else:
      return float(0.0)

  def v_good(self):
    if (self.mid_score > self.titik4 and self.mid_score < self.titik5):
      return round(float(self.mid_score - self.titik4)/(self.titik5 - self.titik4), 2)
    elif (self.mid_score > self.titik5 and self.mid_score < self.titik6):
      return round(float(self.titik6 - self.mid_score)/(self.titik6 - self.titik5), 2)
    else:
      return float(0.0)

  def excellent(self):
    if (self.mid_score > self.titik5 and self.mid_score < self.titik6):
      return round(float(self.mid_score - self.titik5)/(self.titik6 - self.titik5), 2)
    elif (self.mid_score >= self.titik4):
      return float(1.0)
    else:
      return float(0.0)
```
